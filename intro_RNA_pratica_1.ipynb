```python
# Perceptron
# implementação da função AND e OR

import numpy as np

#função de ativação
def degree(u):
  y = []
  for value in u:
    if(value < 0):
      y.append(0)
    else:
      y.append(1)
  return np.array(y)

#implementação do perceptron
def yperceptron(w,b,x, activation_function = degree):
  u = np.dot(w,x) + b                   #implementa a multiplicação de todos os valores de w e x
  y = activation_function(u)
  return y

#declara x
x = np.array([[0, 1, 0, 1], [0, 0, 1, 1]])
```

```python
#AND
w = np.array([0.4, 0.4])
b = -0.6
# y esperado = [0, 0, 0, 1]

y = yperceptron(w,b,x, degree)
print(y)

```

```python
#OR
w = np.array([0.4, 0.4])
b = -0.4
# y esperado = [0, 1, 1, 1]

y = yperceptron(w,b,x, degree)
print(y)
```
